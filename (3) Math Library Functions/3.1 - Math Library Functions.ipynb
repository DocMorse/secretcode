```python
#3.1 Practical Activity 1

"""
LONG COMMENT - Libraries and Modules
-----------------------
In order to keep memory requirements down, python has quite a lightweight
language base, with only a few additional commands.
If we wish to expand on this we can import additional commands
as libraries (also called modules).

Imagine an expansion pack in a video game. It is kind of
the same principle. You only need to install a pack if you actually
need to use it.

In the program below a library (module) called math is imported.
This allows for the programmer to include useful mathematical functions
within a program.

In example below the originalNum variable is passed
into the math.sqrt() function. This will return its square root.
Which is then assigned to the squareRoot variable (Line 46)

Lets look at this in action below.
-------------------------------------
"""

#--------------------Import Libraries------------------

# Importing the math library, so its functions can be used
import math

#--------------------Global Variables------------------

# Initialising a variable called originalNum
# Assigning the integer 0 to it
originalNum = 0

# Initialising a variable called squareRoot
# Assigning the float 0.0 to it
squareRoot = 0.0

#--------------------Main Program----------------------

# Asking the user for a number, casting it to an integer
# and assigning it to the variable originalNum
originalNum = int(input("Please enter a number: "))

# Calculating the square root of the number, with the math.sqrt()
# function, and assigning it to the variable squareRoot
squareRoot = math.sqrt(originalNum)

# Using the round function to round squareRoot to two decimal places
squareRoot = round(squareRoot, 2)

# Outputting the square root of originalNum
print("The square root of", originalNum, "is", squareRoot)
```
