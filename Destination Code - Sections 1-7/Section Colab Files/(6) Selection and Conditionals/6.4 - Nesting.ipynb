```python
#6.4 Practical Activity 1

"""
LONG COMMENT- Conditional Statements
-----------------------
So far all of the conditional statements you have
worked with are linear (then move onto the next part of
the program once a condition has been met (or not met).

Instead imagine a condition only being tested if a previous
condition is true (or false). This is called nesting.

You can see an example of nested conditionals below.
-------------------------------------
"""

#-------------Global Variables------------------

# Initialising a variable called yesNo
# Assigning an empty string value to it.
yesNo = ""

#--------------------Main Program----------------------

# Asking the user to input yes or no in response to the question
# The response is converted to lower case
# It is assigned to the variable yesNo
yesNo = input("Is it sunny out? ").lower()

# Using a nested conditional sequence to recommend
# what the user should wear
if yesNo == "yes":
  print("So it is sunny out")
  yesNo = input("Is it windy? ").lower()

  if yesNo == "yes":
    print("Wear a hoodie")
    print("It might be sunny but a breeze will make it chilly")

  else:
    print("Just wear a t-shirt")
    print("It is sunny so you won't want too many layers")

else:
  print("So it is raining out")
  yesNo = input("Is it windy? ").lower()

  if yesNo == "yes":
    print("Wear a raincoat")
    print("It is too windy for an umbrella")

  else:
    print("Take an umbrella")
    print("If there is no wind, you want full protection from the rain")
```
