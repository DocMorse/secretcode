```python
#6.1 Practical Activity 1

"""
LONG COMMENT- Conditional Statements
-----------------------
So far all of the programs you have written have been simple.
All they have consisted of is inputs, outputs and mathematical
operations.

Now we will investigate how to make decisions in Python.

In order to do this we will be using conditional statements.
One way to identify conditional statements is the fact that
they contain the command words IF, ELIF and ELSE.

-------------------------------------
"""

#-------------Global Variables------------------

# Initialising a variable called qAnswer
# Assigning an empty string value to it.
qAnswer = ""

# Initialising a variable called qCorrect
# Assigning an empty string value to it.
qCorrect = ""

#--------------------Subprograms-----------------------

# Creating a subprogram called answerChecker
# Note that it has two parameters.
# When the subprogram is called a is replaced
# with the user's answer and c is replaced with
# what the correct answer should be.

def answerChecker(a, c):
  # Below is a conditional expression. The double
  # equals sign is used to test if a and c are equal.
  if (a == c):
    print("Correct") # Response if true
  else:
    print("Incorrect") # Response if false

#--------------------Main Program----------------------

# Asking the user to input their answer to a question
# Assigning their answer to the variable qAnswer
qAnswer = input("Is the sky blue? ")

# Assigning the string "yes" to the variable qCorrect
qCorrect = "yes"

# Calling the answerChecker subprogram
# The arguments qAnswer and qCorrect are used as arguments
# to replace the parameters a and c
answerChecker(qAnswer, qCorrect)
```

```python
# 6.1 Practical Activity 6

#------------------Global Variables-----------------

#====> Initialise a variable called memberInput
#====> Assign an empty string to it.


#------------------Main Program-----------------------

#====> Ask the user to input the name of a member of 1 Direction.
#====> convert their input to lower case using .lower()
#====> Assign the input to a variable called memberInput


"""
LONG COMMENT
----------------
Below you can see a branched conditional sequence.

This allows for more than one correct answer using
the ELIF command.

Branched conditional sequences always begin with an IF,
have lots of ELIFS in the middle and end with an ELSE.
----------------
"""

# Creating a branched conditional sequence that tests
# if memberInput is equal to "harry", "niall", "louis",
# "liam", or "zayn".
# It responds with either correct or incorrect
if memberInput == "harry":
  print("Correct! Harry Styles was a member of 1 Direction.")
elif memberInput == "niall":
  print("Correct! Niall Horan was a member of 1 Direction.")
elif memberInput == "louis":
  print("Correct! Louis Tomlinson was a member of 1 Direction.")
elif memberInput == "liam":
  print("Correct! Liam Payne was a member of 1 Direction.")
elif memberInput == "zayn":
  print("Correct! Zayn Malik was a member of 1 Direction.")
else:
  print("Incorrect. That person was not a member of 1 Direction.")

```
