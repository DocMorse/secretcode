```python
#6.5 Practical Activity 8

#-------------Global Variables------------------

# Initialising two numbers to use in the program
num1 = 0
num2 = 0

#-------------Main Program-----------------------

# Asking the user to input a value for each number
# Assigning the user inputs to their respective numbers
num1 = int(input("Enter number 1:"))
num2 = int(input("Enter number 2:"))

# Using a conditional sequence to test
# the relationship between the numbers
if num1 > num2:
    print(num1, "is greater than", num2)
elif num1 < num2:
    print(num1, "is less than", num2)
else:
    print(num1, "is equal to", num2)

"""
Create a flow diagram to represent the above program
You can use either google slides or lucidchart.
https://www.lucidchart.com/
"""
```

```python
#6.5 Practical Activity 9

#-------------Global Variables------------------

# Initialising a variable to assign user answers to
answer = ""

#-------------Main Program-----------------------

# Asking the user to input an answer to a question
# Assigning the user input to the variable answer
answer = input("Have you been to Scotland? ")

if answer == "yes":
    answer = input("Have you been to Stirling? ")
    if answer == "yes":
        print("Me too, I really liked the castle")
    else:
        print("Shame, you should see it next time you go")

else:
    print("Sorry to hear, you should consider visiting")

"""
Create a flow diagram to represent the above program
You can use either google slides or lucidchart.
https://www.lucidchart.com/
"""
```

```python
#6.5 Practical Activity 10

#-------------Global Variables------------------

# Initialising two variables to use in the program
diet = ""
choice = ""

#-------------Main Program-----------------------

# Asking the user their dietary requirements
diet = input("Are you vegetarian (1)Yes (2)No: ")

# Testing their dietary requirements
if diet == "(1)":
  choice = input("Enter Choice (1)Veggie Burger (2)Veggie SpagBol: ")
  if choice == "(1)":
    print("You have ordered a Veggie Burger")
  elif choice == "(2)":
    print("You have ordered a Veggie SpagBol")
  else:
    print("Invalid Choice")

else:
  choice = input("Enter Choice (1)Chicken Burger (2)Chicken Nuggets: ")
  if choice == "(1)":
    print("You have ordered a Chicken Burger")
  elif choice == "(2)":
    print("You have ordered a Chicken Nuggets")
  else:
    print("Invalid Choice")

"""
Create a flow diagram to represent the above program
You can use either google slides or lucidchart.
https://www.lucidchart.com/
"""
```
