```python
#7.1 Practical Activity 1

"""
LONG COMMENT
----------------
Repetition is a programming concept where a sequence of
instructions is repeated until a condition is no longer met.

In order to create repetitive sequences in Python we
use While Loops.

A while loop begins with a condition (similar to IF)
however THERE IS NO Else statement (lots of people mistakenly
add them when they are learning while loops).

At the beginning of a while loop, if the condition is met
the loop is entered. If the condition isn't met, the loop
is skipped and Python moves on to the next section of the code.
--------------------------------
"""

#---------------Global Variables-------------------

# Initialising a variable that will be used
# to store a password entered by the user
password = ""

#---------------Main Program-----------------------

# Asking the user to input what they think is the
# password. This is then assigned to the variable password
password = input("Enter Password: ")

# Using a while loop to test if password is not equal
# to the string on line 39. If it isn't the user is told
# their password is invalid and they are given a chance
# to try again. These steps are repeated until the user
# correctly guesses the password.
while password != "Str0ngP455w0rd":
  print("Invalid Password - Try Again")
  password = input("Enter Password: ")

# Once the user correctly guesses the password the
# condition for the while loop is invalid and the loop ends.
# The program then moves on to the next line, which
# outputs "Password Accepted".
print("Password Accepted")
```
