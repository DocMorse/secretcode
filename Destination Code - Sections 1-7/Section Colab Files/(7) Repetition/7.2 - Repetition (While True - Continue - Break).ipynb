```python
#7.2 Practical Activity 1

""" The code below uses a while loop, however it begins
with the condition while True. This use of the boolean
value True, unrelated to any other values confuses
the computer into infinitely repeating the loop.

This is a similar structure to the forever loop in
Scratch.

The only way a while True loop can be ended is with
the break command.
"""

#---------------------Global Variables------------------------

# Initialising a variable called password
# Assigning an empty string to it
password = ""

#---------------------Main Program----------------------------

# Creating a while True loop. This will repeat infinitely until
# the break command is used
while True:

  # Asking the user to input a password
  # Assigning the user input to password
  password = input("Enter your password: ")

  # Using a conditional statement to test if password == "pass123"
  # If the condition is true, the program outputs "Correct"
  # and the while loop breaks
  # If the condition is false it outputs "incorrect, try again"
  if password == "pass123":
    print("Correct")
    break
  else:
    print("Incorrect, try again")

# Once the while loop has broken a message is output, welcoming the user
# to the system
print("Welcome to the system")
```
