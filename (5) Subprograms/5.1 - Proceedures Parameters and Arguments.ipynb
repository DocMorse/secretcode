```python
#5.1 Practical Activity 1

"""
LONG COMMENT - Subprograms
-----------------------
Subprograms are named, reusable blocks of code.

They allow us to write a block of code once in a program but
reuse it as many times as we want.

On line 33 the subprogram mythicalBeast is defined.
This means that all code indented under this line is
part of the subprogram and will always be run when the
subprogram is called into action.

On line 46 you can see the calling code for this particular
subprogram. By simply typing its name followed by brackets
the whole subprogram will be called into action and be run
as if all of its code existed at this point in the program.

On line 49 the subprogram is called into action again.
Python will then act as if each line of code in the subprogram
is present at this point in the program as well.
-------------------------------------
"""

#-------------Global Variables------------------

# There are no global variables within this program


#-------------Subprograms------------------------
def mythicalBeast():
  size = input("Enter a size like greater, lesser, small, giant: ")
  decor = input("Enter a decoration such as spotted, striped, crested etc : ")
  noise = input("Enter a sound such as screeching, roaring, howling")
  animal = input("Enter the name of an animal e.g. cat or dog")

  beast = size + " " + decor + " " + noise + " " + animal

  print("I went to the woods and saw a", beast)

#--------------------Main Program----------------------

# Calling the subprogram into action
mythicalBeast()

# Calling the subprogram into action for a second time
mythicalBeast()

#====> Run the program and see what happens.
```
