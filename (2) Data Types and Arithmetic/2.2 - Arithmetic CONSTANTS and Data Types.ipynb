```python
#2.2 Practical Activity 1

"""
LONG COMMENT - DATA TYPES AND CASTING
-----------------------
There are four main data types used by Python

1) String: This is used for text and has no
mathematical value. All inputs by default are string

2) Integer: This is used for storing whole numbers.

3) Float (also called Real): This is used for storing
numbers that include decimal values.

4) Boolean: This is used for storing the result of a test
and its value can either be True or False.

In the code below you will see the program asks the user
to input an integer with the value of 6.

As all inputs by default are string data, the user's input
needs to be converted to the integer data type.

We call this process CASTING.
You can see casting on line 47
-------------------------------------

"""

#-------------Global Variables------------------

# Initialising a variable called numberOne and assigning the integer
# 0 to it
numberOne = 0

# Initialising a variable called answer and assigning the integer
# 0 to it.
answer = 0

#--------------------Main Program----------------------

# Asking the user to input a number and assigning it to numberOne
numberOne = input("Enter a number to multiply by 2: ")

# Casting (changing) the value of numberOne to the integer data type.
numberOne = int(numberOne)

# Multiplying numberOne by 2
# Then assigning this to a the variable answer
answer = numberOne * 2

# Outputting the calculation and the answer
print(numberOne, "X 2 =", answer)
```
