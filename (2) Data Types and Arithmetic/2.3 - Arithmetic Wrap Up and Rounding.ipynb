```python
#2.3 Practical Activity 1

"""
LONG COMMENT - ROUNDING
-----------------------
In the program below you will see that
the rounding function has been used.

round() is one of many "built in functions" or
BIFS that Python comes with to help you with
your arithmetic.

You have already used functions when casting
inputs to integers and floats. The difference
with round is how many values (known as "arguments")
we place in the brackets.

Each "argument" must be separated by a comma.

With the round function the first argument is the
number to be rounded and the second argument is the number
of decimal places to round it to.

You can see this on line 56
-------------------------------------
"""

#-------------Global Variables------------------

# Initialising a variable called numberOne
# Assigning the integer 0 to it.
numberOne = 0

# Initialising a variable called answer
# As the answer will contain the result of a division
# I have assigned the float 0.0 to it.
answer = 0.0

#--------------------Main Program----------------------

# Asking the user to input a number and assigning it to numberOne
numberOne = input("Enter a number: ")

# Casting (changing) the value of numberOne to the integer data type.
numberOne = int(numberOne)

# Dividing the number by 731
# Then assigning this to a the variable answer
answer = numberOne / 731

# Outputting the calculation and the answer
print(numberOne, "/ 731 =", answer)
print("But this has too many decimal places")

# Using the round function to limit the answer to two decimal places
answer = round(answer, 2)

# Spacing the output
print("")

# Outputting the answer again but with 2 decimal places
print("Now it only has two decimal places")
print(numberOne, "/ 731 =", answer)
```
