```python
# 4.3 Practical Activity 4

"""
LONG COMMENT - The Random Choice
-----------------------
It is also possible to generate a random choice
from a list of words.

You will see this demonstrated below

Lists are like variables but they can contain
multiple values
-------------------------------------
"""

#--------------------Import Libraries------------------

# Importing the random library
import random

#--------------------Global Variables------------------

# Initialising a list called fruits
# A list is like a variable but it has multiple values
# Assigning multiple values to the list.
fruits = ["apple","banana","pear","orange"]

# Initialising a variable called randomFruit
# Assigining an empty string to it
randomFruit = ""

#--------------------Main Program----------------------

# Using the random.choice() function to select an item
# from the list
# Assigning the value to the randomFruit variable
randomFruit = random.choice(fruits)

# Outputting a message to tell the user what their random fruit is
print("Your random fruit is", randomFruit)
```
